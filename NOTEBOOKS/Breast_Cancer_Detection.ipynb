```python
!pip install opendatasets
import opendatasets as od
od.download("https://www.kaggle.com/datasets/emiliovenegas1/mammography-dataset-from-inbreast-mias-and-ddsm/data")
# {"username":"shivamsukhija19","key":"ca290767041ea50faaaa1f0c2dabead9"}


# Full beginner-friendly pipeline for mammogram classification
# ------------------------------------------------------------
# Put this file next to your "mammography-dataset-from-inbreast-mias-and-ddsm/CLAHE_images" folder
# and update ORIG_DATA_DIR below.

import os
import random
import shutil
from pathlib import Path
from collections import Counter

import numpy as np
import matplotlib.pyplot as plt

import tensorflow as tf
from tensorflow import keras
from sklearn.utils.class_weight import compute_class_weight
from sklearn.metrics import classification_report, confusion_matrix

# ==============
# CONFIG
# ==============
ORIG_DATA_DIR = "/content/mammography-dataset-from-inbreast-mias-and-ddsm/CLAHE_images"
OUT_DIR = "data_split"   # where train/val/test will be created
IMG_SIZE = (224, 224)
BATCH_SIZE = 16
SEED = 42
TEST_RATIO = 0.15
VAL_RATIO = 0.15
TRAIN_RATIO = 1.0 - TEST_RATIO - VAL_RATIO

# ==============
# 1) Split function: create train/val/test folders
# ==============
def create_split_folders(orig_dir, out_dir, train_ratio=0.7, val_ratio=0.15, test_ratio=0.15, seed=42):
    """
    Split images in 'orig_dir' (which contains class subfolders) into
    out_dir/train, out_dir/val, out_dir/test while preserving class folders.
    This function copies files (safe) and uses reproducible shuffling.
    """
    random.seed(seed)
    orig_dir = Path(orig_dir)
    out_dir = Path(out_dir)

    # make output base folders
    for split in ["train", "val", "test"]:
        (out_dir / split).mkdir(parents=True, exist_ok=True)

    # iterate class folders (benign, malignant, normal)
    class_folders = [p for p in orig_dir.iterdir() if p.is_dir()]
    class_folders.sort()  # deterministic order

    for cls_folder in class_folders:
        cls_name = cls_folder.name
        # gather files
        files = [p for p in cls_folder.iterdir() if p.is_file() and p.suffix.lower() in [".png", ".jpg", ".jpeg", ".tif", ".tiff", ".bmp"]]
        files.sort()
        random.shuffle(files)

        n = len(files)
        n_train = int(round(n * train_ratio))
        n_val = int(round(n * val_ratio))
        # ensure sums don't exceed n
        if n_train + n_val > n:
            n_val = max(0, n - n_train)

        train_files = files[:n_train]
        val_files = files[n_train:n_train + n_val]
        test_files = files[n_train + n_val:]

        # create class subfolders in output
        for split_name, split_files in [("train", train_files), ("val", val_files), ("test", test_files)]:
            target_folder = out_dir / split_name / cls_name
            target_folder.mkdir(parents=True, exist_ok=True)
            for src in split_files:
                dst = target_folder / src.name
                # copy file (safe; keeps original untouched)
                shutil.copy(src, dst)

    print(f"Data split created at: {out_dir}")
    # print counts
    for split in ["train", "val", "test"]:
        summary = {}
        split_base = out_dir / split
        for cls in sorted([p.name for p in split_base.iterdir() if p.is_dir()]):
            summary[cls] = len(list((split_base/cls).iterdir()))
        print(split, summary)

# ==============
# 2) Create tf.data datasets using image_dataset_from_directory
# ==============
def load_datasets(base_dir, img_size=(224,224), batch_size=16, seed=42):
    """
    Use Keras helper to create train/val/test datasets.
    Returns train_ds, val_ds, test_ds, class_names
    """
    train_dir = os.path.join(base_dir, "train")
    val_dir   = os.path.join(base_dir, "val")
    test_dir  = os.path.join(base_dir, "test")

    train_ds_raw = tf.keras.utils.image_dataset_from_directory(
        train_dir,
        labels="inferred",
        label_mode="int",
        image_size=img_size,
        batch_size=batch_size,
        shuffle=True,
        seed=seed
    )

    val_ds_raw = tf.keras.utils.image_dataset_from_directory(
        val_dir,
        labels="inferred",
        label_mode="int",
        image_size=img_size,
        batch_size=batch_size,
        shuffle=False
    )

    test_ds_raw = tf.keras.utils.image_dataset_from_directory(
        test_dir,
        labels="inferred",
        label_mode="int",
        image_size=img_size,
        batch_size=batch_size,
        shuffle=False
    )

    # get class names BEFORE prefetch
    class_names = train_ds_raw.class_names
    print("Classes found (index order):", class_names)

    # prefetch to improve speed
    AUTOTUNE = tf.data.AUTOTUNE
    train_ds = train_ds_raw.prefetch(AUTOTUNE)
    val_ds = val_ds_raw.prefetch(AUTOTUNE)
    test_ds = test_ds_raw.prefetch(AUTOTUNE)

    return train_ds, val_ds, test_ds, class_names


# ==============
# 3) Data augmentation (apply only to training)
# ==============
def get_augmentation_layer():
    """
    Return a small augmentation Sequential model that can be used in the tf.data pipeline
    or inside the model. Keep augmentations simple and safe for medical images.
    """
    data_augmentation = keras.Sequential([
        keras.layers.RandomFlip("horizontal"),
        keras.layers.RandomRotation(0.05),
        keras.layers.RandomZoom(0.05),
        keras.layers.RandomContrast(0.1)
    ], name="data_augmentation")
    return data_augmentation

def augment_dataset(dataset, augmentation_layer):
    """
    Apply augmentation to a dataset (assumes batched dataset of (images, labels)).
    Returns a new dataset where images have augmentation applied.
    """
    def apply_augment(images, labels):
        images = augmentation_layer(images, training=True)
        return images, labels

    # map with num_parallel_calls not using lambda
    return dataset.map(lambda x, y: apply_augment(x, y), num_parallel_calls=tf.data.AUTOTUNE)

# ==============
# 4) Compute class weights from train folder files (so imbalance is handled)
# ==============
def compute_weights_from_folder(train_folder):
    """
    Compute class weights using file counts in train_folder.
    Returns a dict {class_index: weight}
    """
    class_names = sorted([d.name for d in Path(train_folder).iterdir() if d.is_dir()])
    labels = []
    for idx, cls in enumerate(class_names):
        cls_folder = Path(train_folder) / cls
        count = len([p for p in cls_folder.iterdir() if p.is_file()])
        labels.extend([idx] * count)

    unique = np.unique(labels)
    weights = compute_class_weight(class_weight="balanced", classes=unique, y=np.array(labels))
    return dict(enumerate(weights))

# ==============
# 5) Build model (EfficientNetB0 backbone with custom head)
# ==============
def build_model(img_size=(224,224), num_classes=3, dropout_rate=0.4):
    """
    Build a model using EfficientNetB0 as backbone.
    Returns model and base_model (so we can unfreeze base later).
    """
    input_shape = (img_size[0], img_size[1], 3)
    # input layer
    inputs = keras.Input(shape=input_shape)

    # preprocess input for EfficientNet
    x = keras.applications.efficientnet.preprocess_input(inputs)

    # base pretrained model (feature extractor)
    base_model = keras.applications.EfficientNetB0(
        include_top=False,
        weights="imagenet",
        input_shape=input_shape
    )
    base_model.trainable = False  # freeze at first

    x = base_model(x, training=False)
    x = keras.layers.GlobalAveragePooling2D()(x)
    x = keras.layers.Dropout(dropout_rate)(x)
    outputs = keras.layers.Dense(num_classes, activation="softmax")(x)

    model = keras.Model(inputs, outputs)
    return model, base_model

# ==============
# 6) Training function: train head, then fine-tune
# ==============
def train_and_finetune(model, base_model, train_ds, val_ds, class_weights,
                       head_epochs=10, finetune_epochs=20, checkpoint_path="best_model.h5"):
    """
    Two-phase training:
      - Phase 1: train top head (base frozen)
      - Phase 2: unfreeze some of base and fine-tune with smaller LR
    """
    # Phase 1: compile and train only top layers
    model.compile(
        optimizer=keras.optimizers.Adam(learning_rate=1e-4),
        loss="sparse_categorical_crossentropy",
        metrics=["accuracy"]
    )

    checkpoint_cb = keras.callbacks.ModelCheckpoint(checkpoint_path, save_best_only=True, monitor="val_loss")
    earlystop_cb = keras.callbacks.EarlyStopping(monitor="val_loss", patience=6, restore_best_weights=True)

    print("Starting phase 1 (train head)...")
    history1 = model.fit(
        train_ds,
        validation_data=val_ds,
        epochs=head_epochs,
        class_weight=class_weights,
        callbacks=[checkpoint_cb, earlystop_cb]
    )

    # Phase 2: unfreeze last layers of base
    base_model.trainable = True
    # Freeze early layers to avoid destroying pretrained features
    for layer in base_model.layers[:100]:
        layer.trainable = False

    model.compile(
        optimizer=keras.optimizers.Adam(learning_rate=1e-5),
        loss="sparse_categorical_crossentropy",
        metrics=["accuracy"]
    )

    print("Starting phase 2 (fine-tune)...")
    history2 = model.fit(
        train_ds,
        validation_data=val_ds,
        epochs=finetune_epochs,
        class_weight=class_weights,
        callbacks=[checkpoint_cb, earlystop_cb]
    )

    return model, (history1, history2)

# ==============
# 7) Evaluation helpers
# ==============
def evaluate_model(model, test_ds, class_names):
    """
    Evaluate on the test set, print classification report and confusion matrix.
    """
    # get y_true and y_pred
    y_true = []
    y_pred = []

    for images, labels in test_ds:
        preds = model.predict(images)
        preds_idx = np.argmax(preds, axis=1)
        y_true.extend(labels.numpy().tolist())
        y_pred.extend(preds_idx.tolist())

    print("Classification report:")
    print(classification_report(y_true, y_pred, target_names=class_names))

    print("Confusion matrix:")
    print(confusion_matrix(y_true, y_pred))

def plot_training(history_tuple):
    """
    Plot accuracy and loss for both phases (history1, history2)
    """
    histories = []
    for h in history_tuple:
        if hasattr(h, "history"):
            histories.append(h.history)

    plt.figure(figsize=(12,4))
    # accuracy
    plt.subplot(1,2,1)
    for i, h in enumerate(histories):
        plt.plot(h["accuracy"], label=f"train_acc_phase{i+1}")
        plt.plot(h["val_accuracy"], label=f"val_acc_phase{i+1}")
    plt.legend()
    plt.title("Accuracy")

    # loss
    plt.subplot(1,2,2)
    for i, h in enumerate(histories):
        plt.plot(h["loss"], label=f"train_loss_phase{i+1}")
        plt.plot(h["val_loss"], label=f"val_loss_phase{i+1}")
    plt.legend()
    plt.title("Loss")
    plt.show()

# ==============
# 8) Single image prediction helper
# ==============
def predict_image(model, image_path, class_names, img_size=(224,224)):
    """
    Load a single image, preprocess it exactly as training, predict and print readable result.
    """
    img = keras.preprocessing.image.load_img(image_path, target_size=img_size)
    x = keras.preprocessing.image.img_to_array(img)
    x = np.expand_dims(x, axis=0)
    x = keras.applications.efficientnet.preprocess_input(x)  # same preprocessing used in model
    preds = model.predict(x)
    idx = np.argmax(preds[0])
    confidence = float(np.max(preds[0]))
    label = class_names[idx]
    print(f"Prediction: {label} (confidence={confidence:.3f})")
    return label, confidence

# ==============
# 9) RUN EVERYTHING (main sequence)
# ==============
if __name__ == "__main__":
    # 1) create split folders (only runs if data_split not exists)
    if not os.path.exists(OUT_DIR):
        create_split_folders(ORIG_DATA_DIR, OUT_DIR, train_ratio=TRAIN_RATIO, val_ratio=VAL_RATIO, test_ratio=TEST_RATIO, seed=SEED)
    else:
        print(f"{OUT_DIR} already exists. Skipping split (delete folder to re-split).")

    # 2) load datasets
    train_ds, val_ds, test_ds, class_names = load_datasets(OUT_DIR, img_size=IMG_SIZE, batch_size=BATCH_SIZE, seed=SEED)

    # 3) augmentation (only to training)
    augmentation_layer = get_augmentation_layer()
    train_ds = augment_dataset(train_ds, augmentation_layer)

    # 4) compute class weights from the train folder on disk
    class_weights = compute_weights_from_folder(os.path.join(OUT_DIR, "train"))
    print("Class weights:", class_weights)

    # 5) build model
    model, base_model = build_model(img_size=IMG_SIZE, num_classes=len(class_names), dropout_rate=0.4)
    model.summary()

    # 6) train and fine-tune
    model, histories = train_and_finetune(model, base_model, train_ds, val_ds, class_weights,
                                          head_epochs=10, finetune_epochs=20, checkpoint_path="best_model.h5")

    # 7) plot training curves
    plot_training(histories)

    # 8) evaluate on test set
    evaluate_model(model, test_ds, class_names)

    # 9) save final model
    model.save("breast_cancer_model_final.h5")
    print("Saved final model to breast_cancer_model_final.h5")

    # 10) quick prediction example (user replace path with their image)
    # example_image = "some_test_image.png"
    # predict_image(model, example_image, class_names)
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from tensorflow import keras

# Load trained model
model = keras.models.load_model("breast_cancer_model_final.h5")

# Class names (same order as training)
class_names = ["benign", "malignant", "normal"]

IMG_SIZE = (224, 224)

def predict_and_show(model, image_path, class_names, img_size=(224,224)):
    # 1) Load image
    orig_img = cv2.imread(image_path)
    if orig_img is None:
        raise ValueError(f"Could not load image at {image_path}")

    # Convert BGR → RGB (for matplotlib)
    orig_img_rgb = cv2.cvtColor(orig_img, cv2.COLOR_BGR2RGB)

    # 2) Preprocess for model
    img = cv2.resize(orig_img, img_size)
    x = keras.preprocessing.image.img_to_array(img)
    x = np.expand_dims(x, axis=0)
    x = keras.applications.efficientnet.preprocess_input(x)

    # 3) Predict
    preds = model.predict(x)
    idx = np.argmax(preds[0])
    confidence = float(np.max(preds[0]))
    label = f"{class_names[idx]} ({confidence:.2f})"

    # 4) Show image with title
    plt.figure(figsize=(6,6))
    plt.imshow(orig_img_rgb)
    plt.axis("off")
    plt.title(label, fontsize=16, color="green")
    plt.show()

    return label, confidence

# 🔍 Example usage
image_path = "/content/breast cancer 2.webp"
predict_and_show(model, image_path, class_names)
```
