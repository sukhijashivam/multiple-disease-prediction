Train Test Split

```
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.2, stratify=Y, random_state=2)
```

# Standardize

```
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

Training the Model

```
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, AdaBoostClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from xgboost import XGBClassifier
```

```
models = {
    "SVM (RBF)": SVC(kernel='rbf', C=2, gamma=0.1),
    "SVM (Linear)": SVC(kernel='linear'),
    "Logistic Regression": LogisticRegression(max_iter=3000),
    "Random Forest": RandomForestClassifier(n_estimators=300),
    "Decision Tree": DecisionTreeClassifier(),
    "KNN": KNeighborsClassifier(),
    "Gradient Boosting": GradientBoostingClassifier(),
    "AdaBoost": AdaBoostClassifier(),
    "XGBoost": XGBClassifier(
        n_estimators=300,
        learning_rate=0.05,
        max_depth=4,
        subsample=0.9,
        colsample_bytree=0.9,
        eval_metric="logloss"
    )
}
```

```
results = {}
trained_models = {}

for name, model in models.items():

    # Models that require scaled data
    if name in ["SVM (RBF)", "Logistic Regression", "KNN"]:
        model.fit(X_train_scaled, Y_train)
        preds = model.predict(X_test_scaled)
    else:
        model.fit(X_train, Y_train)
        preds = model.predict(X_test)

    acc = accuracy_score(Y_test, preds)
    results[name] = acc
    trained_models[name] = model

    print(f"{name}: {acc:.4f}")
```

```
best_model_name = max(results, key=results.get)
best_model = trained_models[best_model_name]

print("\nBest Model:", best_model_name)
print("Accuracy:", results[best_model_name])
```

Model Evaluation

Accuracy Score

```
# accuracy score on the training data
if best_model_name in ["SVM (RBF)", "Logistic Regression", "KNN"]:
    X_train_pred = best_model.predict(X_train_scaled)
else:
    X_train_pred = best_model.predict(X_train)

train_acc = accuracy_score(Y_train, X_train_pred)
print("Training Accuracy of best model:", train_acc)
```

```
# accuracy score on the testing data
if best_model_name in ["SVM (RBF)", "Logistic Regression", "KNN"]:
    X_test_pred = best_model.predict(X_test_scaled)
else:
    X_test_pred = best_model.predict(X_test)

test_acc = accuracy_score(Y_test, X_test_pred)
print("Test Accuracy:", test_acc)
print("\n")
```
